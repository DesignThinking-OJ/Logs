```python
def read_label(label_file_path):
    #从label文件中读出现有的label，为list的形式
    label_file=open(label_file_path)
    readline=label_file.readlines()
    if(len(readline)!=0):
        old_label=readline[0].split()
        if(old_label[-1].isdigit()):
            old_label=old_label[:-1]
    else:
        old_label=[]
    return old_label
    
def write_label(label_file_path,label_list):
    file=open(label_file_path,'w')
    label_list.append(str(len(label_list)))
    file.write(" ".join(label_list));
    file.close()
    
def all_labels(new_file_path_list,new_file_num):
    #可以将新加入的文件标签更新进标签集合中，标签集合就是/Volumes/D/jupyter_notebook/标签/label.txt
    #并且之前的标签顺序不会改变，也就是已经用旧标签标好的文件不需要重新改题向量
    old_label=read_label("/Volumes/D/jupyter_notebook/标签/label.txt")

    new_labels=[]
    for i in range(new_file_num):
#         print(i)
        df=pd.read_excel(new_file_path_list[i])
#         print(new_file_path_list[i])
        for index, row in df.iterrows():
#             print(row["label"])
            if(type(row["label"]!=str)):
                continue
            new_labels+=row["label"].split()
        new_labels=list(set(new_labels))
    all_labels=old_label
    for j in range(len(new_labels)):
        if(new_labels[j] not in all_labels):
            all_labels.append(new_labels[j])
            
    write_label("/Volumes/D/jupyter_notebook/标签/label.txt",all_labels)

    return all_labels

new_file_path_list=['/Volumes/D/jupyter_notebook/标签/primer_8_9_校验二.xlsx','/Volumes/D/jupyter_notebook/标签/primer_11_校验二.xlsx','/Volumes/D/jupyter_notebook/标签/primer12_15_校验二.xlsx']
all_labels(new_file_path_list,3)

```

```python
def tovec(label_file_path,file_path,save_file_path):
    print(file_path,save_file_path)
    df=pd.read_excel(file_path)
    print(df.size)
    labels=read_label(label_file_path)
    labels.insert(0,"id")
#     print(labels)
    vector = pd.DataFrame(columns = labels)
    vector["id"]=df["习题id"]
    vector=vector.fillna(value=0)
    
    label_list=df["label"]
    count=0
#     print(label_list)
    for row in label_list:
        row_label=row.split()
        for i in range(len(row_label)):
            i_index=labels.index(row_label[i])
            vector.iloc[count,i_index]=1
        count+=1
        
    vector.to_csv(save_file_path,index=False,sep=',')
#     print(save_file_path)
    print(vector)
    
# filepath='/Volumes/D/jupyter_notebook/标签/primer12_15_校验二.xlsx'
# label_file_path="/Volumes/D/jupyter_notebook/标签/label.txt"
# tovec(label_file_path,filepath)
file_path_list=['/Volumes/D/jupyter_notebook/标签/primer_8_9_校验二.xlsx','/Volumes/D/jupyter_notebook/标签/primer_11_校验二.xlsx','/Volumes/D/jupyter_notebook/标签/primer12_15_校验二.xlsx']
label_file_path="/Volumes/D/jupyter_notebook/标签/label.txt"
save_file_path_list=['/Volumes/D/jupyter_notebook/标签/primer_8_9_vector.xlsx','/Volumes/D/jupyter_notebook/标签/primer_11_vector.xlsx','/Volumes/D/jupyter_notebook/标签/primer12_15_vector.xlsx']
for i in range(len(file_path_list)):
    tovec(label_file_path,file_path_list[i],save_file_path_list[i])
```
